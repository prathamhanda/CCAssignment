Question 1

```python
paragraph = """Technology is evolving at an incredible pace. From smartphones to artificial intelligence, every aspect of our life is being reshaped. I love exploring new gadgets and understanding how they work. The Internet of Things connects our devices for a smarter lifestyle. It's fascinating to see how innovation drives progress every day."""

# 1. Convert to lowercase and remove punctuation
import re
lowered = paragraph.lower()
no_punct = re.sub(r'[^\w\s]', '', lowered)
print("No punctuation:\n", no_punct)

# 2. Tokenize sentences and words
from nltk.tokenize import sent_tokenize, word_tokenize
sentences = sent_tokenize(paragraph)
words = word_tokenize(paragraph)
print("Sentences:", sentences)
print("Words:", words)

# 3. Compare split() vs word_tokenize()
split_words = paragraph.split()
print("Using split():", split_words)
print("Using word_tokenize():", words)

# 4. Remove stopwords
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words if w.lower() not in stop_words]
print("Without stopwords:", filtered_words)

# 5. Word frequency distribution (excluding stopwords)
from nltk.probability import FreqDist
fdist = FreqDist(filtered_words)
fdist.plot(10, title="Word Frequency")
```

Question 2

```python
# 1. Extract words with only alphabets
alpha_words = re.findall(r'\b[a-zA-Z]+\b', paragraph)
print("Alphabetic words:", alpha_words)

# 2. Remove stopwords
filtered_alpha = [w for w in alpha_words if w.lower() not in stop_words]
print("Filtered:", filtered_alpha)

# 3. Stemming with PorterStemmer
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed = [stemmer.stem(w) for w in filtered_alpha]
print("Stemmed:", stemmed)

# 4. Lemmatization with WordNetLemmatizer
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(w.lower()) for w in filtered_alpha]
print("Lemmatized:", lemmatized)

# 5. Comparison
print("Stemming may reduce words to non-readable roots (e.g., 'connects' → 'connect'), while Lemmatization gives valid words (e.g., 'devices' → 'device').")
```

Question 3

```python
texts = [
    "The iPhone 15 has a stunning display and improved battery life.",
    "Samsung Galaxy delivers top-tier performance and camera quality.",
    "Pixel phones offer pure Android experience with timely updates."
]

# 1. Bag of Words using CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
bow = count_vect.fit_transform(texts)
print("Bag of Words (BoW):")
print(bow.toarray())
print("Feature Names:", count_vect.get_feature_names_out())

# 2. TF-IDF using TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vect = TfidfVectorizer()
tfidf = tfidf_vect.fit_transform(texts)
print("\nTF-IDF Scores:")
print(tfidf.toarray())
print("TF-IDF Feature Names:", tfidf_vect.get_feature_names_out())

# 3. Top 3 keywords from each text
import numpy as np
feature_names = tfidf_vect.get_feature_names_out()
for i, row in enumerate(tfidf.toarray()):
    top3_indices = np.argsort(row)[-3:][::-1]
    print(f"\nTop 3 keywords in text {i+1}:")
    for idx in top3_indices:
        print(f"{feature_names[idx]} (score: {row[idx]:.3f})")
```

Question 4

```python
text1 = """Artificial Intelligence is revolutionizing the tech industry.
It allows machines to mimic human intelligence and perform complex tasks."""
text2 = """Blockchain is a decentralized technology that secures data.
It ensures transparency and immutability in financial transactions."""

# 1. Preprocess and tokenize
from nltk.tokenize import word_tokenize
text1_tokens = set(word_tokenize(text1.lower()))
text2_tokens = set(word_tokenize(text2.lower()))

# 2a. Jaccard Similarity
intersection = text1_tokens.intersection(text2_tokens)
union = text1_tokens.union(text2_tokens)
jaccard_sim = len(intersection) / len(union)
print(f"Jaccard Similarity: {jaccard_sim:.3f}")

# 2b. Cosine Similarity using TF-IDF
from sklearn.metrics.pairwise import cosine_similarity
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform([text1, text2])
cos_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
print(f"Cosine Similarity: {cos_sim[0][0]:.3f}")

# 2c. Analysis
print("Cosine similarity captures the angle between vectors and often gives a better sense of content similarity, especially for longer texts.")
```

```python
# Q5: Sample review
review = "I absolutely love the new headphones! The sound quality is crisp and the battery lasts all day."

# 1. Sentiment using TextBlob
from textblob import TextBlob
blob = TextBlob(review)
print("Polarity:", blob.sentiment.polarity)
print("Subjectivity:", blob.sentiment.subjectivity)

# 2. Classify sentiment
if blob.sentiment.polarity > 0:
    sentiment = "Positive"
elif blob.sentiment.polarity < 0:
    sentiment = "Negative"
else:
    sentiment = "Neutral"
print("Review Sentiment:", sentiment)

# 3. Word Cloud (assuming multiple positive reviews)
from wordcloud import WordCloud
import matplotlib.pyplot as plt

positive_reviews = [
    "Amazing phone with great performance.",
    "Loved the user interface and display quality!",
    "Battery life is super impressive."
]
text = " ".join(positive_reviews)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud for Positive Reviews")
plt.show()
```

```python
# Q6: Text Generation with LSTM (fixed imports for TensorFlow)

training_text = """Technology is advancing faster than ever. With artificial intelligence, we can automate tasks and create smart applications. The integration of IoT devices connects the physical world with the digital one. From smart homes to self-driving cars, innovation is shaping our future. As data grows, so does the need for powerful computing and secure networks."""

# 1. Tokenize using tensorflow.keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
import numpy as np

# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts([training_text])
sequences = tokenizer.texts_to_sequences([training_text])[0]
vocab_size = len(tokenizer.word_index) + 1

# 2. Create input sequences
input_sequences = []
for i in range(1, len(sequences)):
    n_gram_seq = sequences[:i+1]
    input_sequences.append(n_gram_seq)

# Padding
max_len = max([len(seq) for seq in input_sequences])
input_sequences = pad_sequences(input_sequences, maxlen=max_len, padding='pre')
X, y = input_sequences[:, :-1], input_sequences[:, -1]

# 3. Build and train model
model = Sequential()
model.add(Embedding(vocab_size, 10, input_length=max_len-1))
model.add(LSTM(50))
model.add(Dense(vocab_size, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')
model.fit(X, y, epochs=200, verbose=0)

# Generate text
seed_text = "technology"
next_words = 5

for _ in range(next_words):
    token_list = tokenizer.texts_to_sequences([seed_text])[0]
    token_list = pad_sequences([token_list], maxlen=max_len-1, padding='pre')
    predicted = model.predict(token_list, verbose=0).argmax()
    output_word = tokenizer.index_word.get(predicted, '')
    seed_text += " " + output_word

print("Generated text:", seed_text)
```
